```python
import sklearn
import sklearn.tree
import matplotlib.pyplot as plt
import sklearn.linear_model
import pandas as pd
import sklearn.ensemble
import os
import pandas as pd
import tensorflow as tf
```

```python
df = pd.read_csv("/Users/hardiktaneja/Downloads/TREE-ml/train.csv")
```

```python
cols =df.columns
```

```python
# x = df.drop(['Cover_Type','Soil_Type7','Soil_Type15','Soil_Type25','Soil_Type37','Soil_Type40','Soil_Type36'],axis=1)
x = df.drop(['Cover_Type','Id'],axis=1)
y = df["Cover_Type"]
```

```python
x_train,x_test,y_train,y_test = sklearn.model_selection.train_test_split(x,y,test_size=.2)
```

```python
clf_gini = sklearn.tree.DecisionTreeClassifier(criterion="gini",random_state=100,max_depth=3,min_samples_leaf = 5)
clf_gini.fit(x_train,y_train)
clf_gini.score(x_test,y_test)
```

```python
clf = sklearn.ensemble.RandomForestClassifier(random_state=0)
clf.fit(x_train,y_train)
clf.score(x_test,y_test)
```

```python
feature_columns=[]
for key in df.keys():
    if(key == 'Id'):
        continue
    elif(key == 'Cover_Type'):
        continue
    feature_columns.append(tf.contrib.layers.feature_column.real_valued_column(key))
```

```python
Classifier = tf.estimator.DNNClassifier(feature_columns = feature_columns,hidden_units=[10,10],n_classes=8)
```

```python
def f1():
    return dict(x_train),y_train
def f2():
    return dict(x_test),y_test
```

```python
Classifier.train(input_fn = f1,steps= 50000)
eval_result = Classifier.evaluate(input_fn =f2,steps=1)
print(eval_result)
```
